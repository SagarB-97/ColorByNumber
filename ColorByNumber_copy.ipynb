```python
from colorbynumber.main import ColorByNumber
from colorbynumber.config import default_config
from colorbynumber import utils
```

```python
# Image from: https://www.pexels.com/photo/scarlet-macaw-1618424/
# Photographer: Tim Mossholder

image_path = "/Users/skalasib/Desktop/ColorByNumber/Tree.jpeg"
color_list = [
    # Paints
    (251,213,6), # Yellow
    (255,174,107), # Light Orange
    (255,164,54), # Orange
    (245,119,86), # Light red
    (252,137,169), # Pink
    (239,226,192), # Offwhite
    (138,123,103), # Grey

    (153,118,196), # Purple
    (219,190,212), # Light Purple
    (82,160,222), # Light blue
    (99,157,232), # Blue
    (26,194,171), # Bluish Green
    (205,236,183), # Light green
    (85,172,105), # Green
    (137,218,188), # Light green 2
]

# See colorbynumber/config.py for an explanation of configs
config = dict(default_config)
config["denoise_type"] = "fastNlMeansDenoisingColored"
config["denoise_h"] = 10
config["area_perc_threshold"] = 0.0
config["font_size"] = 0.2
config["font_thickness"] = 1
config["debug"] = True

colorbynumber_obj = ColorByNumber(
    image_path = image_path, 
    color_list = color_list,
    # num_colors = 10,
    config = config,
)
numbered_islands = colorbynumber_obj.create_color_by_number()

# utils.show_image(colorbynumber_obj.image)
# utils.show_image(colorbynumber_obj.simplified_image)
# utils.show_image(numbered_islands, cmap = 'grey')
# utils.show_image(colorbynumber_obj.generate_color_legend())
```
